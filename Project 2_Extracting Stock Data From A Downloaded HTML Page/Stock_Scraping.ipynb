<h1>Extracting Stock Data From An HTML Page Using Python Libraries</h1>
<h2>Description</h2>
<p>In this lab we used BeautifulSoup to extract stock data from an HTML Page we downloadeds. We then create a dataframe to display the data</p>

```python
from bs4 import BeautifulSoup 
import pandas as pd
```

For this lab we used <a href = 'https://finance.yahoo.com/gainers/heatmap'>yahoo finance</a> to get our stock data. To avoid retrieving the data from the website I went on to the site and downloaded the html page as an html file.

```python
with open('top_stock_yahoo_finance.html') as html_page:
    soup =BeautifulSoup(html_page, 'html.parser')
 
```

Inother to get the data from the table we need to loop through the table to accomplish this task of data extraction.

```python
# Here we will try an extract the data in the table head and --
# later on use it as the column of a dataframe we are going to create.

headings = list()
table_head = soup.find_all('thead')

for row in table_head:
    cells =row.find_all('th')
    for cell in cells:
        headings.append(cell.text)

# After the extraction of the data in the table head we try --
# and do the same thing for the table body by looping throuh the body and --
# every cell.

data = list()
table_body = soup.find_all('tbody')
for row in table_body:
    cells_data =row.find_all('td')
    for cell_data in cells_data:
        data.append(cell_data.text)

data
```
